```python
%matplotlib inline
import sys
print(sys.version)
import numpy as np
print(np.__version__)
import pandas as pd
print(pd.__version__)
import matplotlib.pyplot as plt
from pandas_datareader import data, wb
from datetime import timedelta
```

```python
altitudes = pd.read_csv('stadium_altitudes.csv')
```

```python
altitudes
```

```python
altitudes.set_index('team', inplace=True)
altitudes.head()
```

```python
stationmap = {'yankees':'KNYC', 'red_sox':'KBOS', 'orioles':'KBWI', 'rays':'KSPG', 'blue_jays':'KYYZ',
             'mets':'KLGA', 'braves':'KATL', 'phillies':'KPHL', 'nationals':'KDCA', 'marlins':'KMIA',
             'white_sox':'KMDW', 'indians':'KBKL', 'tigers':'KDET', 'royals':'KMKC', 'twins':'KMSP',
             'cubs':'KORD', 'reds':'KLUK', 'brewers':'KMKE', 'pirates':'KPIT', 'cardinals':'KCPS',
             'astros':'KEFD', 'angels':'KSLI', 'athletics':'KOAK', 'mariners':'KBFI', 'rangers':'KDAL',
             'd-backs':'KPHX', 'rockies':'KDEN', 'dodgers':'KCQT', 'padres':'KSAN', 'giants':'KSFO'
             }
```

```python
closablepark = {'yankees':False, 'red_sox':False, 'orioles':False, 'rays':True, 'blue_jays':True,
             'mets':False, 'braves':False, 'phillies':False, 'nationals':False, 'marlins':True,
             'white_sox':False, 'indians':False, 'tigers':False, 'royals':False, 'twins':False,
             'cubs':False, 'reds':False, 'brewers':True, 'pirates':False, 'cardinals':False,
             'astros':True, 'angels':False, 'athletics':False, 'mariners':True, 'rangers':False,
             'd-backs':True, 'rockies':False, 'dodgers':False, 'padres':False, 'giants':False
             }
```

```python
teams = pd.DataFrame()
```

```python
teams['team'] = stationmap.keys()
teams['stationID'] = teams['team'].map(lambda x: stationmap[x])
teams['isclosable'] = teams['team'].map(lambda x: closablepark[x])
teams.set_index('team', inplace=True)
teams.head()
```

```python
teams = teams.join(altitudes, how='outer')
```

```python
teams = teams.reset_index()
```

```python
teams
```

```python
teams.to_csv('Home_Data.csv')
```
