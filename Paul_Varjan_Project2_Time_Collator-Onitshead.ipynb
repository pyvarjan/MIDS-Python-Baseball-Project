```python
hw = pd.read_csv('HomeWeatherData.csv')
del hw['Unnamed: 0']
hw.gmt = pd.to_datetime(hw.gmt)
hw['masterweatherID'] = hw['team'] + ' ' + hw['gmt'].apply(lambda x: str(x))
hw.head()
```
