```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', 500)
```

```python
stationmap = {'yankees':'KNYC', 'red_sox':'KBOS', 'orioles':'KBWI', 'rays':'KSPG',
            'mets':'KLGA', 'braves':'KATL', 'phillies':'KPHL', 'nationals':'KDCA', 'marlins':'KMIA',
            'white_sox':'KMDW', 'indians':'KBKL', 'tigers':'KDET', 'royals':'KMKC', 'twins':'KMSP',
            'cubs':'KORD', 'reds':'KLUK', 'brewers':'KMKE', 'pirates':'KPIT', 'cardinals':'KCPS',
            'astros':'KEFD', 'angels':'KSLI', 'athletics':'KOAK', 'mariners':'KBFI', 'rangers':'KDAL',
            'd-backs':'KPHX', 'rockies':'KDEN', 'dodgers':'KCQT', 'padres':'KSAN', 'giants':'KSFO'
            }
```

```python
mlb = pd.read_csv("all-stadiums_short.csv")

# maps stadium to station
mlb['station'] = mlb.stadium
mlb.replace({'station':stationmap}, inplace = True)

# Combines station, day and hour to create new field
mlb['day'] = pd.DataFrame(list(mlb.tfs_zulu.str.split('T')))[0]
mlb['time_full'] = pd.DataFrame(list(mlb.tfs_zulu.str.split('T')))[1]
mlb['hour'] = pd.DataFrame(list(mlb.time_full.str.split(':')))[0]
mlb['new_index'] = mlb.station + mlb.day + mlb.hour

# make new field the index
mlb.set_index('new_index', inplace = True)

# delete intermediary columns
del mlb['time_full']
del mlb['day']
del mlb['hour']
del mlb['Unnamed: 0']
```

```python
# exact same process as with pitch data

weather = pd.read_csv('HomeWeatherData.csv')
weather['day'] = pd.DataFrame(list(weather.gmt.str.split(' ')))[0]
weather['time_full'] = pd.DataFrame(list(weather.gmt.str.split(' ')))[1]
weather['hour'] = pd.DataFrame(list(weather.time_full.str.split(':')))[0]
weather['new_index'] = weather.stationID + weather.day + weather.hour
#weather.set_index('new_index', inplace = True)
del weather['time_full']
del weather['day']
del weather['hour']
del weather['Unnamed: 0']
del weather['stationID']
```

```python
weather_mean = pd.DataFrame(weather.groupby('new_index').mean())
```

```python
merged = mlb.join(weather_mean)
```
